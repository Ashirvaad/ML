```python
import numpy as np
def euclidean_distance(point1,point2):
  point1=np.array(point1)
  point2=np.array(point2)
  return np.sqrt(np.sum((point1-point2)**2))
point1=[1,2,3]
point2=[4,5,6]
distance=euclidean_distance(point1,point2)
print(f"Euclidean distance between {point1} and {point2} is {distance}")
```

```python
import numpy as np
def manhattan_distance(point1,point2):
  point1=np.array(point1)
  point2=np.array(point2)
  distance=np.sum(np.abs(point1-point2))
  return distance
point1=[1,2,3]
point2=[4,5,6]
distance=manhattan_distance(point1,point2)
print(f'Manhattan distance between {point1} and {point2} is {distance}')
```

```python
import numpy as np
def cosine_distance(vector1,vector2):
  vector1=np.array(vector1)
  vector2=np.array(vector2)
  dot_product=np.dot(vector1,vector2)
  norm_vetctor1=np.linalg.norm(vector1)
  norm_vector2=np.linalg.norm(vector2)
  cosine_similarity=dot_product/(norm_vetctor1*norm_vector2)
  cosine_distance=1-cosine_similarity
  return cosine_similarity
vector1=[1,2,3]
vector2=[4,5,6]
distance=cosine_distance(vector1,vector2)
print(f'Cosine distance between {vector1} and {vector2} is {distance}')
```
