```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import datasets

dataset=datasets.load_iris()
iris=pd.DataFrame(dataset.data,columns=dataset.feature_names)
iris['species']=dataset.target
iris.head()
print(iris)

#Scatterplot
plt.figure()
plt.subplot(1,2,1),sns.scatterplot(x=iris['sepal length (cm)'] ,
y=iris['sepal width (cm)'],hue=iris['species'],palette=['red','green','orange'])
plt.subplot(1,2,2),sns.scatterplot(x=iris['petal length (cm)'] ,
y=iris['petal width (cm)'],hue=iris['species'],palette=['red','green','orange'])
plt.title('Comparision of different species wrt sepal length and sepal width')
print(" ")
print(" ")

#Boxplot
plt.figure()
plt.subplot(1,2,1),sns.boxplot(x=iris['species'],y=iris['sepal length (cm)'],orient='v')
plt.subplot(1,2,2),sns.boxplot(x=iris['species'],y=iris['sepal width (cm)'],orient='v')
plt.title('Sepal length distribution for different species')
plt.show()
plt.subplot(1,2,1),sns.boxplot(x=iris['species'],y=iris['petal length (cm)'],orient='v')
plt.subplot(1,2,2),sns.boxplot(x=iris['species'],y=iris['petal width (cm)'],orient='v')
plt.title('Petal length distribution for different species')
plt.show()

#Barchart
plt.figure()
plt.subplot(1,2,1)
plt.bar(iris['species'],iris['sepal length (cm)'],color='green')
plt.title('Barchart Sepal Length')
plt.subplot(1,2,2)
plt.bar(iris['species'],iris['sepal width (cm)'],color='Yellow')
plt.title('Barchart Sepal Width')
plt.show()

plt.subplot(1,2,1)
plt.bar(iris['species'],iris['petal length (cm)'],color='orange')
plt.title('Barchart Petal Length')
plt.subplot(1,2,2)
plt.bar(iris['species'],iris['petal width (cm)'],color='red')
plt.title('BarChart Petal Width')
plt.show()

#Linechart
plt.figure()
plt.title('Linechart')
plt.plot(iris['sepal length (cm)'],label='Sepal Length',color='blue')
plt.plot(iris['sepal width (cm)'],label='Sepal Width',color='Yellow')
plt.xlabel('Number of Samples')
plt.legend()

plt.show()
plt.plot(iris['petal length (cm)'],label='Petal Length',color='green')
plt.plot(iris['petal width (cm)'],label='Petal Width',color='red')
plt.xlabel('Number of Samples')
plt.legend()
plt.show()

```
