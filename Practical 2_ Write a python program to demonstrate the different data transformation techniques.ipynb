***Filtering***

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
print("Head of dataset:")
df.head()
print(df)

element_1st_col_5th_row = df.iloc[4, 0]
print("\nElement on the 1st column and 5th row of Dataset:")
print(element_1st_col_5th_row)

elements_3_to_5_cols = df.iloc[:, 2:5]
print("\nElements from 3rd column to 5th column:")
print(elements_3_to_5_cols)

elements_2nd_3rd_rows = df.iloc[1:3, :]
print("\nElements from 2nd and 3rd row of all columns:")
print(elements_2nd_3rd_rows)

sepal_width_lt_3_5 = df[df['sepal width (cm)'] < 3.5]
print("\nRecords from the dataset having sepal width < 3.5:")
print(sepal_width_lt_3_5)

sepal_len_gt_4_9_petal_len_lte_1_4 = df[(df['sepal length (cm)'] > 4.9) & (df['petal length (cm)'] <= 1.4)]
print("\nRecords from the dataset having sepal length > 4.9 and petal length <= 1.4:")
print(sepal_len_gt_4_9_petal_len_lte_1_4)
```

***Feature Scaling***

```python
from sklearn.datasets import load_diabetes
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Load the diabetes dataset
diabetes = load_diabetes()
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)

print("Original Dataset:")
print(df.head())

# Select columns 's1' and 's2' for scaling
x = df[['s1', 's2']]

# Initialize the MinMaxScaler
scalar = MinMaxScaler()

# Fit and transform the selected columns
scaled_x = scalar.fit_transform(x)

# Create a DataFrame with scaled data
scaled_df = pd.DataFrame(data=scaled_x, columns=['s1', 's2'])

print("\nScaled Dataset:")
print(scaled_df)
```
