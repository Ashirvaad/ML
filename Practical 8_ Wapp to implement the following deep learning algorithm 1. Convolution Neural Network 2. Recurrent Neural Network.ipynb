RNN

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Flatten, Dense, MaxPooling2D
from tensorflow.keras.utils import to_categorical
from sklearn.datasets import load_iris

# Load the Iris dataset
dataset = load_iris()
X = dataset.data
Y = dataset.target

# One-hot encode the target labels
encoder = OneHotEncoder(sparse=False)
Y_encoded = encoder.fit_transform(Y.reshape(-1, 1))

# Reshape X to fit the CNN input (samples, height, width, channels)
# Here, we can treat each sample as a 2D "image" of size 2x2
X_reshaped = X.reshape(-1, 2, 2, 1)

# Split the dataset
x_train, x_test, y_train, y_test = train_test_split(X_reshaped, Y_encoded, random_state=50, test_size=0.25)

# Build the CNN model
model = Sequential([
    Conv2D(32, kernel_size=(2, 2), activation='relu', input_shape=(2, 2, 1)),
    MaxPooling2D(pool_size=(1, 1)),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(3, activation='softmax')  # 3 classes for iris species
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(x_train, y_train, epochs=50, batch_size=5, verbose=1)

# Evaluate the model
loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
print(f"Test accuracy: {accuracy * 100:.2f}%")
```

RNN

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
from sklearn.datasets import load_iris

# Load the Iris dataset
dataset = load_iris()
X = dataset.data
Y = dataset.target

# One-hot encode the target labels
encoder = OneHotEncoder(sparse=False)
Y_encoded = encoder.fit_transform(Y.reshape(-1, 1))

# Reshape X to fit RNN input (samples, timesteps, features)
# Here, we can treat each sample as a sequence of 2 timesteps with 2 features
X_reshaped = X.reshape(-1, 2, 2)  # 2 timesteps, 2 features

# Split the dataset
x_train, x_test, y_train, y_test = train_test_split(X_reshaped, Y_encoded, random_state=50, test_size=0.25)

# Build the RNN model
model = Sequential([
    SimpleRNN(32, input_shape=(2, 2), return_sequences=False),
    Dense(3, activation='softmax')  # 3 classes for iris species
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(x_train, y_train, epochs=50, batch_size=5, verbose=1)

# Evaluate the model
loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
print(f"Test accuracy: {accuracy * 100:.2f}%")
```
